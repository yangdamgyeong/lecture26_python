```python
filename='./sample.txt'

with open(filename, encoding='utf-8') as f_read:
    content = f_read.read()
print(content)
```

```python
with open(filename) as f_read:
    lines = f_read.readline()

for line in lines:
    print(line.strip())
```

```python
with open(filename) as f_read:
    while True:
        line = f_read.readline()
        if line:
            print(line)
        else:
            break
```

```python
fruits = ['사과', '토마토', '바나나']
outfile = 'fruits.txt'
with open(outfile, mode ='w', encoding ='utf-8') as f_write:
    for fruit in fruits:
        f_write.write(fruit+'\n')
```

```python
with open(outfile, mode ='w', encoding ='utf-8') as f_write:
    f_write.writelines(fruits)
```

```python
with open(outfile, mode ='w', encoding ='utf-8') as f_write:
    f_write.write('\n'.join(fruits))
```

```python
with open(filename, mode='rb') as f_read:
    content = f_read.read()
print(content)
print(content.decode('utf-8'))
```

### 학습 되돌아보기(2)

```python
f = open('pyzenex.txt', 'wt')
f.write('파이썬 철학\n')
f.write('아름다움이 추한 것보다 낫다.\n')
f.write('명시적인 것이 암묵적인 것보다 낫다.\n')
f.write('단순함이 복잡한 것보다 낫다.\n')
f.close()
```

```python
f = open('pyzenex.txt')
print(f.read())
f.close()
```

```python
f = open('pyzenex.txt')
for line in f.readline():
    print(line, end='')
f.close
```

```python
f = open('pyzenex.txt')
for line in f:
    print(line, end='')
f.close()
```

```python
f = open('pyzenex.txt')
while True:
    line = f.readline()
    if not line:
        break
    print(line, end='')
f.close()
```

### 학습 되돌아보기 (3)

```python
with open('cp.txt', 'wt') as f:
    f.write('freshman\n')
    f.write('sophomore\n')
    f.write('junior\n')
    f.write('senior\n')

with open('cp.txt', 'rt') as f:
    for grade, name in enumerate(f, start=1):
        print(f'{grade}학년 {name}')
```
