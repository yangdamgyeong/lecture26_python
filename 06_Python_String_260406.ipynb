```python
#[실습] 
#알파벳 두 개를 입력 받아서 두 개 사이에 몇 개의 알파벳이 있는지 계산

a, b = input().split()

a_num = ord(a)
b_num = ord(b)

if a_num > b_num:
    count = a_num - b_num - 1
else:
    count = b_num - a_num - 1

if count < 0:
    count = 0

print(count)
```

### 모스 부호 번역기

```python
# 모스 부호 리스트 만들기
morse_list = [ '.-', '-...', '-.-.', '-..', '.', '..-.', '--.', '....', '..', '.---', '-.-', '.-..', '--', '-.', '---', '.--.', '--.-', '.-.', '...', '-','..-', '...-', '.--', '-..-', '-.--', '--..' ]
text = input().upper()  

result = []

for i in text:
    if i != ' ':  
        idx = ord(i) - 65  # A=0, B=1 ...
        result.append(morse_list[idx])

print(' '.join(result))
# 알파벳 문자열 입력 만들기
# 모스부호가 나와야
```
