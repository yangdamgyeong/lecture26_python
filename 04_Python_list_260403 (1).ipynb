### 리스트의 항목에 접근하기 : 슬라이싱(slicing)

```python
test_list = [1, 2, 3, 4, 5, 6, 7, 8]

print(test_list[0:5])
print(test_list[:5])
print(test_list[1::3])
print(test_list[:5:2])
print(test_list[::3])
print(test_list[:])
print(test_list[::])

print(test_list[-1:-7:-2])
print(test_list[1:-1:])
print(test_list[-3::-1])
print(test_list[::-1])
print(test_list[::-2])
```

### 리스트 재배열 : reverse, sort, sorted

```python
a_list = ['one', 'two', 'three', 'four']
print('원본 데이터',a_list)
a_list.reverse()
print('역순 후 데이터', a_list)

a_list.sort()
print('정렬 후 데이터', a_list)

a_list = ['one', 'two', 'three', 'four']
b_list = sorted(a_list)
print('원본 데이터', a_list)
print('정렬 후 데이터',b_list)

c_list = sorted(a_list, key=len)
print(c_list)
```

```python
scores = [95, 45, 21, 74, 55]
print(sorted(scores, reverse=True))

print(scores)
scores.sort(reverse=True)
print(scores)
```

### 리스트에 항목 추가하기 : append, insert, extend

```python

N = int(input("숫자 :"))
num_list = []
for i in range(1, N + 1):
    num_list.append(i)
print(num_list)
```

```python

N = int(input("숫자 :"))
num_list = [i for i in range(1, N + 1)]
print(num_list)

# expression
num_list = [i * 2 for i in range(1, N + 1)]
print(num_list)

# if문 사용
N = int(input("숫자 :"))
num_list = [i for i in range(1, N + 1) if i % 2 == 0]
print(num_list)
```

### 리스트 탐색하기 : in index count len max min sum

```python
test_list = [1, 2, 3, 1, 2, 3]

target = 3
print(target in test_list)
print(test_list.index(target))
print(test_list.count(target))

target = 4
print(target in test_list)
print(test_list.count(target))
print(test_list.index(target))
```

```python
scores = [85, 92, 78, 90, 100]

count = len(scores)
print(f"시험을 본 학생 수: {count}명")

p_score = max(scores)
print(f"최고 점수: {scores}점")

lowest_score = min(scores)
print(f"최저 점수: {lowest_score}점")

total_sum = sum(scores)
print(f"전체 점수 합계: {total_sum}점")

average = total_sum / count
print(f"평균 점수: {average}점")
```
