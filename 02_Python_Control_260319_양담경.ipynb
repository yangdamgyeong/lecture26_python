```python
# while문 연습 : 입력값에 따라 반복 종료하기
# 사용자로부터 숫자를 하나씩 입력 받아서 총합과 평균을 출력
# 사용자가 0를 입력하면 숫자 입력을 멈춤

sum = count = 0
num = int(input("숫자  입력 (0 : 끝) :"))
while num != 0:
    sum += num
    count += 1
    num = int(input("숫자 입력 (0 : 끝) :"))

# 0 으로 나누기를 하면 ZeroDivisionError 발생 -> 예외처리
if count != 0:
    average = sum / count
else:
    average = 0
    
print(f">> 총합 : {sum}, 평균 : {average:.2f}")
```

```python
# 무한 루프와 분기문 break 활용
sum = count = 0
while True:
    num = int(input("숫자 입력 :"))
    if num == 0:
        break
    sum += num
    count += 1

# 0 으로 나누기를 하면 ZeroDivisionError 발생 -> 예외처리
if count != 0:
    average = sum / count
else:
    average = 0
    
print(f">> 총합 : {sum}, 평균 : {average:.2f}")    
```
