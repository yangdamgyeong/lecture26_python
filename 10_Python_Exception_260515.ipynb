```python
# 예외처리
try:
    num_list = [2, 6, 34, 65, 11]
    index = int(input('인덱스 입력: '))
    print(num_list[index])
except (ValueError, IndexError):
    print(f'0부터 {len(num_list)-1}까지의 정수를 입력하세요.')
except Exception as e:
    print(f'예외가 발생했습니다 : {e} -> {type(e)}')
```

```python
while True:
    try:
        number = int(input("양의 정수 입력 : "))
        if number <= 0:
            raise ValueError
    except ValueError:
        print("양의 정수를 입력하세요.")
    else:
        sum = 0
        for i in range(1, number + 1):
            sum += i
        print(f'>>1부터 {number}까지의 합 : {sum}')
        break
```

### 학습 내용 되돌아보기(2)

```python
try:
    m = int(input('월(month) 입력 > '))
    if not (1 <= m <= 12):
        raise Exception('1 - 12 사이의 정수를 입력하세요.')
except Exception as e:
    print('예외 발생 : {}'.format(e))
else:
    print('%d월: 잘 입력했습니다. ' % m)
```

### 학습 내용 되돌아보기(3)

```python
try:
    def divide(x, y):
        return x / y
    x = int(input('숫자를 입력하세요 :'))
    y = int(input('숫자를 입력하세요 :'))

    result = divide(x, y)

except Exception as e:
    print('0으로는 나눌 수 없습니다')
else:
    print('결과 : ', result)
```

### 학습 내용 되돌아보기(4)

```python
try:
    uyear = {1:'freshman', 2:'sophomore', 3:'junior', 4:'senior'}
    u = int(input('대학교 몇 학년이지요? '))

    if not (1 <= u <= 4):
        raise Exception ('예외 발생 이유: 1~4 정수를 입력하세요.')
    print(f'{u}학년: {uyear[u]}')
except Exception as e:
    print(f'예외 발생 이름: {e}')

else:
    print('예외 처리가 잘되는 군요!')

finally:
       print('끝났어요')
```
