```python
# 구구단 출력 함수
gugudan(5)

def gugudan(num):
    print(f'[구구단 {num}단 ]')
    for i in range(1,10):
        print(f'{num} x {i} = {num*i}')
```

```python
number = int(input('단 입력'))
gugudan(number)
```

```python
# 지역변수, 전역변수

sum = 10
def test(num):
    global sum
    sum += num

test(5)
print(sum)
```

```python
stack = []
capacity = 5

def isFull():
    if len(stack) == capacity:
        return True
    else:
        return False

def push(data):
    if isFull():
        print('Stack이 차 있어서 더 이상 추가할 수가 없습니다.')
    else:
        stack.append(data)

def pop():
    if isEmpty():
        return -1
    else:
        return stack.pop()

def isEmpty():
    return len(stack) == 0
    
def peek():
    if isEmpty():
        return -1
    else:
        return stack[-1]

#=======================================================
print(f'[[ 정수형 스택 연산 실습(용량 : {capacity})]]')
print("=================================================")
print(" 1. Push      2. Pop      3. Peek       0. Exit ")
print("=================================================")

while True:


    menu = int(input('> 메뉴 선택 : '))
    if menu == 0:
        break
        
    elif menu == 1:
        data = int(input('데이터 입력 : '))
        push(data)
    elif menu == 2:
        data = pop()
        if data == -1:
            print('Stack이 비어 있습니다.')
        else:
            print(f'가져온 데이터: {data}')
            
    elif menu == 3:
        data = peek()
        if data != -1:
            print(f'가져올 데이터: {data}')
   
    print('> 현재 스택 상태', stack )

print('[[ 정수형 스택 연산 실습 종료]]')    
```
