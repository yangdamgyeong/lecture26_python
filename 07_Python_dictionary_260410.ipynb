### 딕셔너리(dictionary)

```python
dic = {'name': 'pey', 'phone': '010-9999-1234', 'birth': '1118'}
a = {1: 'hi'}
a = {'a': [1, 2, 3]}

print(dic ['name'])
```

```python
prices = {'아메리카노':2000, '카푸치노':3500, '라떼':3000, '녹차':2500}
print(prices)
```

```python
#체스판의 말 위치 저장

chess_board = {
    (0, 0): "white Rook",
    (0, 1): "White Knight",
    (7, 7): "Black Rook"
}

print(chess_board)
```

```python
# 커피메뉴

coffee_menu = {'아메리카노':2000, '라떼':3000, '녹차':2500}
print(coffee_menu ['라떼'])

print(coffee_menu.get('카푸치노', '없는 메뉴'))

print('카푸치노' in coffee_menu)
```

```python
coffee_menu = {'아메리카노':2000, '라떼':3000, '녹차':2500}

dict_items = coffee_menu.items()
dict_keys = coffee_menu.keys()
dict_values = coffee_menu.values()

print(dict_items)
print(dict_keys)
print(dict_values)

for key in dict_keys:
    print(key)

for item in dict_items:
    print(item)

for value in dict_values:
    print(value)
```

```python
coffee_menu = {'아메리카노':2000, '라떼':3000, '녹차':2500}
coffee_menu['라떼'] = 3500
coffee_menu['카푸치노'] = 4000
print(coffee_menu)
```

```python
coffee_menu = {'아메리카노':2000, '라떼':3000, '녹차':2500}

data = coffee_menu.pop('라떼')
print(data)

del coffee_menu['아메리카노']
print(coffee_menu)

coffee_menu.clear()
print(coffee_menu)
```
