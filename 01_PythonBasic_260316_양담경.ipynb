### 02. 대입연산자
* 두 변수의 값을 교환하는 프로그램 (swap)

```python
a = 10
b = 20
print(a, b)
temp = a
a = b
b = temp
print(a, b)
```

```python
a = b = c = 7
print(a, b, c)
```

```python
a, b, c = 7, 11, 19
print(a, b, c)
```

```python
a = 15
b = 25
print(a, b)
a, b = b, a
print(a, b)
```

```python
a = 10>20
print(a)
```

```python
# 원의 넓이 구하기

# 반지름 입력받기
radius = int(input("원의 반지름  :'))
# 면적 계산하기
pi = 3.14159265
area = radius * radius * pi
# 면적 출력하기
print(f'반지름이 {radius}인 원의 면적은 {area:.2f}입니다.')]
```
