### 학습되돌아보기 (2)

```python
import math
import random
```

```python
def getarea(r):
    area = r * r * math.pi
    return area
```

```python
radius = random.random() * 10
```

```python
radius = round(radius, 2)
```

```python
area = getarea(radius)
```

```python
print('원의 반지름:', radius)
```

```python
print('원주율 pi: ', math.pi)
```

```python
print(f'반지름 {radius}인 원의 면적은 {area:.2f}')
```

### 학습 되돌아보기(3)

```python
import math
```

```python
import statistics
```

```python
if __name__ == '__main__':
    print('1! =', math.factorial(1))
    print('6! =', math.factorial(6))
    print('11! =', math.factorial(11))
    print('16! =', math.factorial(16))

    print()
    
```

```python
st = [80, 99, 77, 65, 92, 74, 82]
print(st)
```

```python
print('중앙값: %.2f' % statistics.median(st))
print('평균: %.2f' % statistics.mean(st))
print('분산: %.2f' % statistics.variance(st))
print('표준편차: %.2f' % statistics.stdev(st))
```
