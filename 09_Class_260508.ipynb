### 파이썬 클래스

```python
class Circle:
    pi = 3.14
    def __init__(self,radius):
        self.radius = radius
    def area(self):
        return self.radius ** 2 * Circle.pi
        

c = Circle(2.5)
print(c.area())
```

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    def area(self):
        return self.width * self.height

rec = Rectangle(2.1, 3.4)
print(rec.area())
```

```python
class Account:
    def __init__(self):
        self.balance = 0
    def deposit(self, amount):
        self.balance += amount
    def withdraw(self, amount):
        self.balance -= amount

a1 = Account()
a1.deposit(100)
print(a1.balance)
```

```python
class Account:
    def __init__(self):
        self.__balance = 0
    def deposit(self, amount):
        self.__balance += amount
    def withdraw(self, amount):
        self.__balance -= amount
    def get_balance(self):
        return self.__balance

a2 = Account()
a2.deposit(50)
print(a2.get_balance())
```
