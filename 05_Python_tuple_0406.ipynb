### 튜플(tuple)

```python
t1 = (1, 2, 3)
t2 = ('a', 'b', ('ab', 'cd'))
t3 = ('tuple', (1, 2), [3, 4])
t4 = ()
t5 = (1,)
t6 = 1, 2, 3
t7 = 1,

print(t1)
print(t2)
print(t3)
print(t6)
x, y, z = t6
print(x, y, z)
```

### 튜플 다루기

```python
t = (10, 20, 30, 40)

# 튜플 항목에 접근 : indexing, slicing, for... in

print(t[0])
print(t[1:3])

for x in t:
    print(x)

# 튜플 탐색 : in 연산자, index(), count()
print(20 in t)
print(t.index(20))
print(t.count(20))

# 튜플 관련 정보 : len(), max(), min(), sum()
print(len(t))
print(max(t))
print(min(t))
print(sum(t))
```

```python
# 새로운 튜플 생성 : sorted(), +(연결), *(반복)

t = (3, 1, 2)

print(tuple(sorted(t)))

t2 = t + (4, 5)
print(t2)

t3 = t * 2
print(t3)

del t
```

### 튜플의 활용

```python
# 두 변수의 값 바꾸기

a = 10
b = 20
a, b = b, a
print(a, b)
```

```python
# 함수에서 여러 개의 값 반환

def add_and_mul(a, b):
    return a+b, a*b

result = add_and_mul(3, 5)
print(result)
```
