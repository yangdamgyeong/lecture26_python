```python
# boj 11478
S = input().strip()
result = set()

for i in range(len(S)):  # 부분 문자열의 시작 index
    for j in range(i, len(S)): # 부분 문자열의 stop index
        sub_str = S[i:j+1]
        result.add(sub_str)

print(len(result))
```
